## REGRESION LINEAL 200 DATOS PRACTICA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("practica.csv")
```

```python
df
```

```python
%matplotlib inline 
plt.plot(df,'b.')
plt.show()
```

```python
plt.plot(df['Altura (pulgadas)'], df['Peso (Libras)'],'b.')
plt.xlabel("Altura Pulgadas")
plt.ylabel("Peso en Libras")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# ConstrucciÃ³n del modelo y ajuste de la funciÃ³n de hipÃ³tesis
lin_reg = LinearRegression()
lin_reg.fit(df['Altura (pulgadas)'].values.reshape(-1,1),(df['Peso (Libras)'].values))
```

```python
# ParÃ¡metro Theta 0 
lin_reg.intercept_
```

```python
# ParÃ¡metro Theta 1
lin_reg.coef_
```

```python
# PredicciÃ³n para el valor mÃ­nimo y mÃ¡ximo del conjunto de datos de entrenamiento
x_min_mx = np.array([[df["Altura (pulgadas)"].min()],[df["Altura (pulgadas)"].max()]])
y_train_pred = lin_reg.predict(x_min_mx)
```

```python
# RepresentaciÃ³n grÃ¡fica de la funciÃ³n hipÃ³tesis generada
plt.plot(x_min_mx, y_train_pred,"g-")
plt.plot(df["Altura (pulgadas)"], df["Peso (Libras)"], "b.")
plt.xlabel("Altura (pulgadas)")
plt.ylabel("Peso (Libras)")
plt.show()
```

```python
x_new = np.array([[79]]) 
#PredicciÃ³n del costo que tendrÃ­a el incidente
Peso = lin_reg.predict(x_new)
print("El peso serÃ­a de: ", int(Peso[0]))
```

```python
plt.plot(df["Altura (pulgadas)"], df["Peso (Libras)"], "b.")
plt.plot(x_min_mx, y_train_pred, "g-")
plt.plot(x_new, Peso, "rx")
plt.xlabel("Altura (pulgadas)")
plt.ylabel("Peso (Libras)")
plt.show()
```
